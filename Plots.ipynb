**Import statements**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Creation of dataframe with the best scores of the classifiers**

```python
df = pd.DataFrame.from_dict({'MLP':98.95,'SVC':98.86,'Random Forest':99.57,'Decision Tree':97.26,'Gaussian Process':96.87,'KNN':97.35,'AdaBoost':99.62,'Voting':99.48},orient='index')
```

```python
df = df.rename(columns={0:'score'}) #rename the column

df = df.sort_values('score') #sort score values
```

**Plot**

```python
df.plot.barh(legend=None,figsize=(10,7))
plt.xlim(96,100)
plt.ylabel('Classifier')
plt.xlabel('Validation Accuracy Score (AUC)')
plt.title('Comparing Classifiers')
#plt.xticks(rotation=45)
plt.show()
```

The AdaBoost had the best score and the Gaussian Process had the worst score.
